```python
#Task-3 Filter the data separately for the state of California,Texas and Montana and save then in 3 different dataframe
#Sort the three dataframe in increasing order of sales amount and display them separately:

California=df[df['State']=='California']
Texas=df[df['State']=='Texas']
Montana=df[df['State']=='Montana']
```

```python
#Task-4 Calculate the sum of sales(six months) for the states specified in task-3 and display the state-wise total sales.
#sample output:Total sales in state of California:11,000:

print('Total sales of California:',df[df['State']=='California']['Sales'].sum())
print('Total sales of Texas:',df[df['State']=='Texas']['Sales'].sum())
print('Total sales of Montana:',df[df['State']=='Montana']['Sales'].sum())

#Interpretation:
#Under this task I has printed the state-wise sum of sales for six months.
```

```python
#Task-5 Display the month with the highest and lowest monthly sale amount for the stores, in the state of California:

df[df['State']=='California'][['Sales','Month','Store']].max()
```

```python
df[df['State']=='California'][['Sales','Month','Store']].min()

#Interpretation:
#Under this task I has displayed the month with the highest(max) and lowest(min) sale amount for stores in the state of California.
```

```python
#Task-6 Create and display a new dataframe consisting of state-wise total sales:

al=df[df['State']=='Alaska']['Sales'].sum()
tl=df[df['State']=='Texas']['Sales'].sum()
cali=df[df['State']=='California']['Sales'].sum()
mon=df[df['State']=='Montana']['Sales'].sum()
new=df[df['State']=='New Mexico']['Sales'].sum()
print(al,tl,cali,mon,new)
```

```python
merge=pd.DataFrame({'State':['Alaska','Texas','California','Montana','New Mexico'],'Sale in thousand':(al,tl,cali,mon,new)})
merge

#Interpretation:
#Under this task I has used the sum function to sum all sales amount
#and then I has converted into dataframe and created a new column name=state,sale in thousand.
```

```python
#Task-7 Sort the rows in the dataframe created in task-6 by the sale amount. Save the dataframe as a total_sale.csvfile in
#your local system.Check your current working directory to see if the file is saved correctly:
sorted=merge.sort_values(by='Sale in thousand')
sorted

sorted.to_csv('Total_sale.csv file')

#Interpretation:
#Under this task I has sorted and saved into new file name is(total_sale.csv file) and checked the sorted details is in the file.Yes, it is in the file. 
```
