```python
import pandas as pd
import numpy as np
```

```python
#Task-1 Create a panda series variable to assign a student_id from 1001-1040:
pan_ser=np.array(np.arange(1001,1041))
student_id=pd.Series(pan_ser)
student_id   

#Interpretation:
#Displayed the student_id from 1001-1040.
```

```python
#Task-2 Store the series in two separate panda series variable and display them:
maths_marks=[35, 86, 63, 90, 50, 63, 15, 72, 15, 54, 72, 83, 58, 70, 92, 40, 59, 63, 53, 59, 55, 85, 18, 83, 67, 85, 46, 59, 77, 71, 77, 56, 79, 61, 82, 76, 60, 52, 88, 78]
physics_marks=[85, 94, 63, 59, 76, 64, 87, 55, 51, 71, 54, 94, 33, 72, 75, 67, 85, 15, 50, 50, 19, 55, 59, 65, 76, 13, 57, 70, 51, 94, 55, 50, 71, 91, 75, 51, 90, 70, 87, 56]

x=pd.Series(data=maths_marks,index=student_id)
x

#Interpretation:
#Displayed the Series maths_marks with the index of student_id.
```

```python
y=pd.Series(physics_marks,index=student_id)
y

#Interpretation:
#Displayed the Series maths_marks with the index of student_id.
```

```python
#Task-3 Display the sum of score in maths and physics:
new_ser=pd.Series(x+y)
new_ser   

#Interpretation:
#created and displayed panda series done a sum of scores of physics and mathematics,along with student_id.
```

```python
#Task-4  Find the students who is qualified .The qualified cut off =120:
qualify=new_ser[new_ser>=120]
qualify

#Interpretation:
#Listed down the qualified marks =120 ,along with student id.
```

```python
#Task-5
add_math_phy=pd.DataFrame(x+y)
add_math_phy
```

```python
#Task-5
changing_name=pd.DataFrame(new_ser[new_ser>=120],columns=['round 1 qualifiers'])
changing_name 

#Interpretation:
#Under this task I has identified the round 1 qualifier using the series and converted into dataframe.
```

```python
#Task-6 We have to find round2 qualifiers so that the maths score and physics score>=50 and even sum of score will be >=120.
round2=x[(x>=50) & (y>=50) & (new_ser >=120)]
round2
round2_qualifier=pd.DataFrame(round2,columns=['round 2 qualifiers'])
print(round2_qualifier)

#Interpretation:
#Under this task I used series and converted into dataframe so that I has identified the round2 qualifiersThus it included total score.
```

```python
#Task-7 Find the count of number of qualified for round-2
round2_qualifier=pd.DataFrame(round2,columns=['round 2 qualifiers'])
print(round2_qualifier)
print('count of qualifier in round2=',len(round2_qualifier))

#Interpretation:
#Under this task I has used len function to find the count of the students qualified for round-2.
```
